```python
import telebot
import threading
import time
import random
import json
import os

TOKEN = "7503868390:AAHoIWkv4RXD2zC8_DXoEchIz75y2ayCGL8"
bot = telebot.TeleBot(TOKEN)

DATA_FILE = "raft_survival_save.json"
SAVE_INTERVAL = 60  # —Å–µ–∫—É–Ω–¥ –º–µ–∂–¥—É –∞–≤—Ç–æ—Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ–º

lock = threading.Lock()

# ========= –î–ê–ù–ù–´–ï ===========

ITEMS = {
    "wood": "ü™µ –î–µ—Ä–µ–≤–æ",
    "stone": "ü™® –ö–∞–º–µ–Ω—å",
    "fiber": "üåø –í–æ–ª–æ–∫–Ω–æ",
    "fish": "üêü –†—ã–±–∞",
    "meat": "ü•© –ú—è—Å–æ",
    "cooked_meat": "üçñ –ü—Ä–∏–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω–æ–µ –º—è—Å–æ",
    "rope": "ü™¢ –í–µ—Ä—ë–≤–∫–∞",
    "boat": "üõ∂ –õ–æ–¥–∫–∞",
    "alien_artifact": "üëΩ –ê—Ä—Ç–µ—Ñ–∞–∫—Ç –ø—Ä–∏—à–µ–ª—å—Ü–µ–≤",
    "chest": "üì¶ –°—É–Ω–¥—É–∫",
}

RECIPES = {
    "rope": {"name": "ü™¢ –í–µ—Ä—ë–≤–∫–∞", "level": 1, "requires": {"fiber": 3}},
    "cooked_meat": {"name": "üçñ –ü—Ä–∏–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω–æ–µ –º—è—Å–æ", "level": 2, "requires": {"meat": 2}},
    "boat": {"name": "üõ∂ –õ–æ–¥–∫–∞", "level": 3, "requires": {"wood": 10, "rope": 2, "stone": 5}},
    "alien_artifact": {"name": "üëΩ –ê—Ä—Ç–µ—Ñ–∞–∫—Ç –ø—Ä–∏—à–µ–ª—å—Ü–µ–≤", "level": 5, "requires": {"stone": 20, "rope": 5, "fish": 10}},
    "chest": {"name": "üì¶ –°—É–Ω–¥—É–∫", "level": 4, "requires": {"wood": 15, "rope": 3}},
}

CHARACTERS = {
    "novice": {"name": "–ù–æ–≤–∏—á–æ–∫", "skill": "–ù–∞—á–∏–Ω–∞—é—â–∏–π —Ä—ã–±–∞–∫", "price": 0},
    "fisher": {"name": "–†—ã–±–∞–∫", "skill": "–õ—É—á—à–µ –ª–æ–≤–∏—Ç —Ä—ã–±—É", "price": 10},
    "hunter": {"name": "–û—Ö–æ—Ç–Ω–∏–∫", "skill": "–£–ª—É—á—à–µ–Ω–Ω—ã–π –∫—Ä—é–∫", "price": 15},
    "alien_hunter": {"name": "–ò—Å—Ç—Ä–µ–±–∏—Ç–µ–ª—å –ø—Ä–∏—à–µ–ª—å—Ü–µ–≤", "skill": "–ó–∞—â–∏—Ç–∞ –æ—Ç –ø—Ä–∏—à–µ–ª—å—Ü–µ–≤", "price": 25},
}

# ============================

def load_data():
    global users
    if os.path.exists(DATA_FILE):
        with open(DATA_FILE, "r", encoding="utf-8") as f:
            users = json.load(f)
    else:
        users = {}

def save_data_loop():
    while True:
        time.sleep(SAVE_INTERVAL)
        with lock:
            with open(DATA_FILE, "w", encoding="utf-8") as f:
                json.dump(users, f, ensure_ascii=False, indent=4)

load_data()
threading.Thread(target=save_data_loop, daemon=True).start()

def init_user(user_id):
    with lock:
        user_id = str(user_id)
        if user_id not in users:
            users[user_id] = {
                "inventory": {"wood": 5, "stone": 3, "fiber": 4, "meat": 0, "fish": 0, "rope": 0, "boat": 0, "cooked_meat": 0, "alien_artifact": 0, "chest": 0},
                "level": 1,
                "exp": 0,
                "hunger": 100,
                "gems": 0,
                "title": "–ù–æ–≤–∏—á–æ–∫",
                "characters": ["novice"],
                "current_char": "novice",
                "last_active": time.time()
            }

def hunger_decrease_loop():
    while True:
        time.sleep(1800)
        with lock:
            for uid in users:
                users[uid]["hunger"] = max(0, users[uid]["hunger"] - 5)

threading.Thread(target=hunger_decrease_loop, daemon=True).start()

def main_menu():
    markup = telebot.types.ReplyKeyboardMarkup(resize_keyboard=True)
    markup.row("üé£ –õ–æ–≤–∏—Ç—å —Ä—ã–±—É", "ü™ù –ë—Ä–æ—Å–∏—Ç—å –∫—Ä—é–∫")
    markup.row("üî® –í–µ—Ä—Å—Ç–∞–∫", "üì¶ –ò–Ω–≤–µ–Ω—Ç–∞—Ä—å")
    markup.row("üçΩ –ü–æ–µ—Å—Ç—å", "üçñ –ì–æ–ª–æ–¥")
    markup.row("üìä –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞", "üèù –û—Å—Ç—Ä–æ–≤")
    markup.row("üéÅ –°–æ–±—ã—Ç–∏—è", "üíé –ú–∞–≥–∞–∑–∏–Ω")
    markup.row("üë§ –ü–µ—Ä—Å–æ–Ω–∞–∂–∏")
    return markup

def crafting_menu(user_id):
    markup = telebot.types.InlineKeyboardMarkup()
    user = users[str(user_id)]
    for key, rcp in RECIPES.items():
        can, reason = can_craft(user, key)
        txt = f"{rcp['name']} {'‚úÖ' if can else '‚ùå'}"
        markup.add(telebot.types.InlineKeyboardButton(txt, callback_data=f"craft_{key}"))
    markup.add(telebot.types.InlineKeyboardButton("‚¨ÖÔ∏è –ù–∞–∑–∞–¥", callback_data="back_main"))
    return markup

def can_craft(user, key):
    r = RECIPES[key]
    if user["level"] < r["level"]:
        return False, f"–¢—Ä–µ–±—É–µ—Ç—Å—è —É—Ä–æ–≤–µ–Ω—å {r['level']}"
    for item, qty in r["requires"].items():
        if user["inventory"].get(item, 0) < qty:
            return False, f"–ù–µ —Ö–≤–∞—Ç–∞–µ—Ç {ITEMS[item]} x{qty}"
    return True, "–ú–æ–∂–Ω–æ"

def craft_item(user, key):
    r = RECIPES[key]
    for item, qty in r["requires"].items():
        user["inventory"][item] -= qty
    user["inventory"][key] = user["inventory"].get(key, 0) + 1
    user["exp"] += 10
    need = user["level"] * 50
    if user["exp"] >= need:
        user["level"] += 1
        user["exp"] -= need
        return True
    return False

@bot.message_handler(commands=["start"])
def cmd_start(m):
    uid = m.from_user.id
    init_user(uid)
    user = users[str(uid)]
    bot.send_message(m.chat.id,
                     f"üåä –ü—Ä–∏–≤–µ—Ç, {m.from_user.first_name}!\n"
                     f"–£—Ä–æ–≤–µ–Ω—å: {user['level']}, –≥–æ–ª–æ–¥: {user['hunger']}/100\n"
                     "–î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å –≤ –∏–≥—Ä—É –≤—ã–∂–∏–≤–∞–Ω–∏—è –Ω–∞ –ø–ª–æ—Ç—É!",
                     reply_markup=main_menu())

@bot.message_handler(func=lambda m: True)
def text_handler(m):
    uid = str(m.from_user.id)
    init_user(uid)
    txt = m.text

    if txt == "üé£ –õ–æ–≤–∏—Ç—å —Ä—ã–±—É":
        fish_event(m)
    elif txt == "ü™ù –ë—Ä–æ—Å–∏—Ç—å –∫—Ä—é–∫":
        hook_event(m)
    elif txt == "üî® –í–µ—Ä—Å—Ç–∞–∫":
        bot.send_message(m.chat.id, "–í—ã–±–µ—Ä–∏ —Ä–µ—Ü–µ–ø—Ç:", reply_markup=crafting_menu(uid))
    elif txt == "üì¶ –ò–Ω–≤–µ–Ω—Ç–∞—Ä—å":
        show_inventory(m)
    elif txt == "üçΩ –ü–æ–µ—Å—Ç—å":
        eat_food(m)
    elif txt == "üçñ –ì–æ–ª–æ–¥":
        bot.send_message(m.chat.id, f"üçñ –¢–≤–æ–π –≥–æ–ª–æ–¥: {users[uid]['hunger']}/100")
    elif txt == "üìä –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞":
        show_stats(m)
    elif txt == "üèù –û—Å—Ç—Ä–æ–≤":
        island_event(m)
    elif txt == "üéÅ –°–æ–±—ã—Ç–∏—è":
        random_event(m)
    elif txt == "üíé –ú–∞–≥–∞–∑–∏–Ω":
        shop_menu(m)
    elif txt == "üë§ –ü–µ—Ä—Å–æ–Ω–∞–∂–∏":
        characters_menu(m)
    else:
        bot.send_message(m.chat.id, "‚ùì –ù–µ–∏–∑–≤–µ—Å—Ç–Ω–∞—è –∫–æ–º–∞–Ω–¥–∞, –∏—Å–ø–æ–ª—å–∑—É–π –º–µ–Ω—é.")

@bot.callback_query_handler(func=lambda call: True)
def callback_handler(call):
    uid = str(call.from_user.id)
    init_user(uid)
    data = call.data
    user = users[uid]

    if data == "back_main":
        bot.edit_message_text("‚¨ÖÔ∏è –ì–ª–∞–≤–Ω–æ–µ –º–µ–Ω—é", call.message.chat.id, call.message.message_id)
        bot.send_message(call.message.chat.id, "–ì–ª–∞–≤–Ω–æ–µ –º–µ–Ω—é:", reply_markup=main_menu())
        return

    if data.startswith("craft_"):
        item = data.split("_")[1]
        can, reason = can_craft(user, item)
        if can:
            leveled_up = craft_item(user, item)
            bot.answer_callback_query(call.id, f"‚úÖ –°–æ–∑–¥–∞–Ω–æ {RECIPES[item]['name']}!")
            if leveled_up:
                bot.send_message(call.message.chat.id, f"üéâ –£—Ä–æ–≤–µ–Ω—å –ø–æ–≤—ã—à–µ–Ω! –¢–µ–ø–µ—Ä—å —É —Ç–µ–±—è —É—Ä–æ–≤–µ–Ω—å {user['level']}")
            bot.edit_message_text("–í—ã–±–µ—Ä–∏ —Ä–µ—Ü–µ–ø—Ç:", call.message.chat.id, call.message.message_id,
                                  reply_markup=crafting_menu(uid))
        else:
            bot.answer_callback_query(call.id, f"‚ùå {reason}", show_alert=True)
        return

    if data.startswith("buychar_"):
        ch = data.split("_")[1]
        if ch in user["characters"]:
            bot.answer_callback_query(call.id, "–ü–µ—Ä—Å–æ–Ω–∞–∂ —É–∂–µ —É —Ç–µ–±—è –µ—Å—Ç—å", show_alert=True)
        else:
            price = CHARACTERS[ch]["price"]
            if user["gems"] >= price:
                user["gems"] -= price
                user["characters"].append(ch)
                bot.answer_callback_query(call.id, f"–ö—É–ø–ª–µ–Ω –ø–µ—Ä—Å–æ–Ω–∞–∂ {CHARACTERS[ch]['name']}!")
            else:
                bot.answer_callback_query(call.id, "–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –≥–µ–º–æ–≤", show_alert=True)
        return

    if data.startswith("selectchar_"):
        ch = data.split("_")[1]
        if ch in user["characters"]:
            user["current_char"] = ch
            bot.answer_callback_query(call.id, f"–í—ã–±—Ä–∞–Ω –ø–µ—Ä—Å–æ–Ω–∞–∂ {CHARACTERS[ch]['name']}")
        else:
            bot.answer_callback_query(call.id, "–£ —Ç–µ–±—è –Ω–µ—Ç —ç—Ç–æ–≥–æ –ø–µ—Ä—Å–æ–Ω–∞–∂–∞", show_alert=True)
        return

def fish_event(m):
    uid = str(m.from_user.id)
    user = users[uid]
    base_chance = 0.6
    if user["current_char"] == "fisher":
        base_chance += 0.2
    if random.random() < base_chance:
        user["inventory"]["fish"] = user["inventory"].get("fish", 0) + 1
        bot.send_message(m.chat.id, "üé£ –ü–æ–π–º–∞–ª —Ä—ã–±—É!")
    else:
        bot.send_message(m.chat.id, "üé£ –†—ã–±–∞ –Ω–µ –∫–ª—é—ë—Ç.")

def hook_event(m):
    uid = str(m.from_user.id)
    user = users[uid]
    loot_pool = ["wood", "stone", "fiber", None, None]
    if user["current_char"] == "hunter":
        loot_pool.append("meat")
    loot = random.choice(loot_pool)
    if loot:
        user["inventory"][loot] = user["inventory"].get(loot, 0) + 1
        bot.send_message(m.chat.id, f"ü™ù –ö—Ä—é–∫ –≤—ã—Ç–∞—â–∏–ª {ITEMS[loot]}!")
    else:
        bot.send_message(m.chat.id, "ü™ù –ö—Ä—é–∫ –Ω–∏—á–µ–≥–æ –Ω–µ –ø–æ–π–º–∞–ª.")

def show_inventory(m):
    uid = str(m.from_user.id)
    user = users[uid]
    text = "üì¶ –ò–Ω–≤–µ–Ω—Ç–∞—Ä—å:\n"
    for k,v in user["inventory"].items():
        if v > 0:
            text += f"{ITEMS[k]}: {v}\n"
    bot.send_message(m.chat.id, text)

def eat_food(m):
    uid = str(m.from_user.id)
    user = users[uid]
    if user["inventory"].get("cooked_meat", 0) > 0:
        user["inventory"]["cooked_meat"] -= 1
        user["hunger"] = min(100, user["hunger"] + 30)
        bot.send_message(m.chat.id, "üçñ –¢—ã —Å—ä–µ–ª –ø—Ä–∏–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω–æ–µ –º—è—Å–æ. –ì–æ–ª–æ–¥ —É–º–µ–Ω—å—à–∏–ª—Å—è!")
    elif user["inventory"].get("fish", 0) > 0:
        user["inventory"]["fish"] -= 1
        user["hunger"] = min(100, user["hunger"] + 15)
        bot.send_message(m.chat.id, "üêü –¢—ã —Å—ä–µ–ª —Å—ã—Ä—É—é —Ä—ã–±—É. –ì–æ–ª–æ–¥ —É–º–µ–Ω—å—à–∏–ª—Å—è, –Ω–æ –Ω–µ —Ç–∞–∫ —Å–∏–ª—å–Ω–æ.")
    else:
        bot.send_message(m.chat.id, "üò¢ –£ —Ç–µ–±—è –Ω–µ—Ç –µ–¥—ã, —á—Ç–æ–±—ã –ø–æ–µ—Å—Ç—å!")

def show_stats(m):
    uid = str(m.from_user.id)
    user = users[uid]
    c = CHARACTERS[user["current_char"]]
    bot.send_message(m.chat.id,
                     f"üìä –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:\n"
                     f"–£—Ä–æ–≤–µ–Ω—å: {user['level']}\n"
                     f"–û–ø—ã—Ç: {user['exp']}\n"
                     f"–ì–æ–ª–æ–¥: {user['hunger']}/100\n"
                     f"–ì–µ–º—ã: {user['gems']}\n"
                     f"–¢–µ–∫—É—â–∏–π –ø–µ—Ä—Å–æ–Ω–∞–∂: {c['name']} ‚Äî {c['skill']}")

def island_event(m):
    uid = str(m.from_user.id)
    user = users[uid]
    events = [
        ("üå¥ –¢—ã —Å–æ–±—Ä–∞–ª –∫–æ–∫–æ—Å—ã!", {"fiber": 2}),
        ("ü¶ç –ù–∞–ø–∞–ª–∞ –æ–±–µ–∑—å—è–Ω–∞! –ü–æ—Ç–µ—Ä—è–ª –≥–æ–ª–æ–¥ –∏ —Ä–µ—Å—É—Ä—Å—ã.", None),
        ("üîç –ü—É—Å—Ç–æ, –Ω–∏—á–µ–≥–æ –Ω–µ –Ω–∞–π–¥–µ–Ω–æ.", None),
        ("üî• –ü–æ–∂–∞—Ä! –ü–æ—Ç–µ—Ä—è–ª —Ä–µ—Å—É—Ä—Å—ã.", None)
    ]
    e = random.choice(events)
    text, rewards = e
    if rewards:
        for k,v in rewards.items():
            user["inventory"][k] = user["inventory"].get(k,0)+v
    else:
        user["hunger"] = max(0, user["hunger"]-20)
        inv_items = [k for k,v in user["inventory"].items() if v>0]
        if inv_items:
            lost = random.choice(inv_items)
            user["inventory"][lost] = max(0, user["inventory"][lost]-1)
    bot.send_message(m.chat.id, text)

def random_event(m):
    uid = str(m.from_user.id)
    user = users[uid]
    events = [
        ("üëΩ –ü—Ä–∏—à–µ–ª—å—Ü—ã –∞—Ç–∞–∫—É—é—Ç! –ü–æ—Ç–µ—Ä—è–ª –≥–æ–ª–æ–¥ –∏ –ø–æ–ª—É—á–∏–ª –∞—Ä—Ç–µ—Ñ–∞–∫—Ç.", -15, {"alien_artifact":1}),
        ("üåßÔ∏è –î–æ–∂–¥—å. –ì–æ–ª–æ–¥ —É–º–µ–Ω—å—à–∏–ª—Å—è –Ω–µ–º–Ω–æ–≥–æ.", -5, None),
        ("üåà –†–∞–¥—É–≥–∞! –ü–æ–ª—É—á–∏–ª –≥–µ–º—ã.", 0, None),
        ("üì¶ –¢—ã –Ω–∞—à—ë–ª —Å—É–Ω–¥—É–∫ —Å –≥–µ–º–∞–º–∏!", 0, {"gems": random.randint(1,5)})
    ]
    e = random.choice(events)
    text, hunger_delta, reward = e
    user["hunger"] = max(0, user["hunger"] + hunger_delta)
    if reward:
        for k,v in reward.items():
            if k == "gems":
                user["gems"] = user.get("gems",0) + v
            else:
                user["inventory"][k] = user["inventory"].get(k,0) + v
    bot.send_message(m.chat.id, text)

def shop_menu(m):
    uid = str(m.from_user.id)
    user = users[uid]
    markup = telebot.types.InlineKeyboardMarkup()
    for k,c in CHARACTERS.items():
        if k in user["characters"]:
            text = f"{c['name']} (–£–∂–µ –∫—É–ø–ª–µ–Ω)"
            markup.add(telebot.types.InlineKeyboardButton(text, callback_data=f"selectchar_{k}"))
        else:
            text = f"{c['name']} - {c['price']} üíé"
            markup.add(telebot.types.InlineKeyboardButton(text, callback_data=f"buychar_{k}"))
    markup.add(telebot.types.InlineKeyboardButton("‚¨ÖÔ∏è –ù–∞–∑–∞–¥", callback_data="back_main"))
    bot.send_message(m.chat.id, "üíé –ú–∞–≥–∞–∑–∏–Ω –ø–µ—Ä—Å–æ–Ω–∞–∂–µ–π", reply_markup=markup)

def characters_menu(m):
    uid = str(m.from_user.id)
    user = users[uid]
    text = "üë§ –¢–≤–æ–∏ –ø–µ—Ä—Å–æ–Ω–∞–∂–∏:\n"
    for c in user["characters"]:
        char = CHARACTERS[c]
        sel = " (–í—ã–±—Ä–∞–Ω)" if c == user["current_char"] else ""
        text += f"{char['name']} ‚Äî {char['skill']}{sel}\n"
    bot.send_message(m.chat.id, text)

if __name__ == "__main__":
    print("–ë–æ—Ç –∑–∞–ø—É—â–µ–Ω!")
    bot.polling(none_stop=True)
```
